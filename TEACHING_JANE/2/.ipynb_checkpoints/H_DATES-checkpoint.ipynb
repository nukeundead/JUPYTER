# Сдать решение задачи H - Сортировка дат
* Полный балл: 150
* Штраф за посылку: 1
* Ограничение времени: 1 с
* Ограничение реального времени: 1 с
* Ограничение памяти: 64М

Необходимо отсортировать набор дат в хронологическом порядке по возрастанию.
Дата состоит из дня, месяца, года и времени с точностью до минут.
Названия месяцев английские, с заглавной буквы.
Диапазон времени 00:00 - 23:59

---
# Формат входных данных
В первой строке число N - количество записей. В последующих N строках даты в формате "ЧислоМесяца НазваниеМесяца Год Час:Минута". Для записи минут используется два знака, т.е. с лидирующим нулём, если количество минут меньше 10.
# Формат выходных данных
N строк - отсортированные записи.

---
# Код

```python
import random


# ОПРЕДЕЛЯЕМ КОЛИЧЕСТВО ДАТ
N = random.randint(3, 7)
print(N)

months_dict = {'January' : 0, 'February' : 1, 'March' : 2, 'April' : 3, 'May' : 4, 'June' : 5, 'July' : 6, 'August' : 7, 'September' : 8, 'October' : 9, 'November' : 10, 'December' : 11}
months = list(months_dict.keys())


# ФУНКЦИЯ ДОБАВЛЕНИЯ ЛИДИРУЮЩЕГО НУЛЯ
def zero_prefix(num):
    if num < 10:
        num = '0' + str(num)
    else:
        num = num
    return num


# ПУЗЫРЬКОВАЯ СОРТИРОВКА, МОДИФИЦИРОВАННАЯ ДЛЯ СОРТИРОВКИ МАССИВОВ ПО ЗАДАННОМУ ЭЛЕМЕНТУ МАССИВА
def bSort(array, pos):
    # ОПРЕДЕЛЯЕМ ДЛИНУ МАССИВА
    l = len(array)
    # ВНЕШНИЙ ЦИКЛ, ЧИСЛО ПРОХОДОВ N-1
    for i in range(l):
        # ВНУТРЕННИЙ ЦИКЛ, N-i-1 ПРОХОДОВ
        for j in range(0, l-i-1):
            # МЕНЯЕМ ЭЛЕМЕНТЫ МЕСТАМИ
            if array[j][pos] > array[j+1][pos]:
                temp = array[j]
                array[j] = array[j+1]
                array[j+1] = temp
    return array


# ГЕНЕРАЦИЯ ДАТ
dates = []
for i in range(N):
    day = random.randint(1, 28)
    month = random.randint(0, 11)
    year = random.randint(1920, 2020)
    hour = zero_prefix(random.randint(0, 23))
    minute = zero_prefix(random.randint(0, 59))

    date = f'{day} {months[month]} {year} {hour}:{minute}'
    dates.append(date)
    print(date)

print()

# ПАРСИМ ВХОДНУЮ СТРОКУ И ПРИВОДИМ ДАННЫЕ К ЦИФРОВОМУ ФОРМАТУ
parse_list = []
i = 0
for date in dates:
    parse_list.append(date.split(' '))
    parse_list[i][0] = int(parse_list[i][0]) # ДНИ
    parse_list[i][1] = int(months_dict[parse_list[i][1]]) # МЕСЯЦЫ
    parse_list[i][2] = int(parse_list[i][2]) # ГОДЫ
    hm = parse_list[i][3].split(':') # [ЧАСЫ, МИНУТЫ]
    parse_list[i][3] = int(hm[0])  # ЧАСЫ
    parse_list[i].append(int(hm[1])) # МИНУТЫ
    print(parse_list[i])
    i += 1

# ОПРЕДЕЛЯЕМ ПОРЯДОК СОРТИРОВКИ - СЕКУНДЫ, ЧАСЫ, ДНИ, МЕСЯЦЫ, ГОДЫ
sort_order = [4, 3, 0, 1, 2]
for i in sort_order:
    parse_list = bSort(parse_list, i)
parse_list

print()

# ВОЗВРАТ К УКАЗАННОМУ ФОРМАТУ ЗАПИСИ
for date in parse_list:
    date = f'{date[0]} {months[date[1]]} {date[2]} {zero_prefix(date[3])}:{zero_prefix(date[4])}'
    print(date)
```
