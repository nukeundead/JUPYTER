```python
import pandas as pd
import os
from datetime import datetime, date, time
```

```python
path = r'C:\Users\LGN\Desktop\РАСЧЕТЫ\ДЗЕН\zen_2020-10-31_dumpof_transactions_1604156325.csv'
```

```python
df = pd.read_csv(path, sep=';', usecols=['date', 'categoryName', 'outcome'])
df
```

```python
s = df['categoryName'].str.split(' / ', expand=True)
df['КАТЕГОРИЯ'] = s[0]
df['ПОДКАТЕГОРИЯ'] = s[1]
df
```

```python
df.columns = ['ДАТА', 'СПЛИТ', 'СУММА', 'КАТЕГОРИЯ', 'ПОДКАТЕГОРИЯ']
df = df.drop('СПЛИТ', axis=1)
df = df.fillna('-')
df['ПОДКАТЕГОРИЯ'] = df['ПОДКАТЕГОРИЯ'].str.upper()
df = df[df['СУММА'] != 0]
df = df.reindex(columns=['КАТЕГОРИЯ', 'ПОДКАТЕГОРИЯ', 'СУММА', 'ДАТА'])
df
```

```python
df['ДАТА'] = pd.to_datetime(df['ДАТА'])
df
```

```python
for index, row in df.iterrows():
    if index == 0:
        week_days = []
    week_days.append(datetime.isoweekday(row['ДАТА']))
df['ДЕНЬ НЕДЕЛИ'] = week_days
df
```

```python
pvt = df.pivot_table(values='СУММА',
                     index='ДАТА',
                     columns='КАТЕГОРИЯ',
                     aggfunc='sum',
                     fill_value=0,
                     margins=False,
                     dropna=True)
pvt
```

```python
pvt = pvt.reset_index()
pvt
```
