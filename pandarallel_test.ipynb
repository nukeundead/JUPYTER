```python
import pandas as pd
import requests

from tqdm import tqdm
tqdm.pandas()

from pandarallel import pandarallel
pandarallel.initialize(progress_bar=True)
from bs4 import BeautifulSoup #beautifulsoup4
```

```python
def address_requester(message_id):
    r = requests.get(f'https://gorod.mos.ru/index.php?message={message_id}')
    soup = BeautifulSoup(r.text, 'lxml')
    address = soup.title.text.replace(' – Наш город Москва', '')
    return address
```
